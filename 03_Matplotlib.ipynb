```python
# Importing necessary libraries
import matplotlib.pyplot as plt  # For data visualization
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical operations
```

```python
# Creating an array of integers from 0 to 9 (inclusive)
x = np.arange(0, 10)

# Generating a new array where each element is twice the corresponding x value
y = 2 * x

# Printing both arrays
print("x values:", x)
print("y values:", y)
```

```python
# Line plot

plt.plot(x, y)  # Plotting the x and y values as a simple line graph
plt.show()  # Displaying the plot
```

```python
# Editing more figure parameters

plt.plot(x, y)  # Plotting the x and y values as a line graph

plt.xlabel("Priyanka's and Arya's X-Axis")  # Setting label for the x-axis
plt.ylabel("Priyanka's and Arya's Y-Axis")  # Setting label for the y-axis
plt.title("Priyanka's and Arya's Plot")  # Assigning a title to the plot
```

```python
# Exporting a plot

plt.plot(x, y)  # Plotting the x and y values as a line graph
plt.savefig('example.png')  # Saving the plot as an image file named 'example.png'
```

```python
# Editing more figure parameters

plt.plot(x, y)  # Plotting the x and y values as a line graph

plt.xlabel("Priyanka's and Arya's X-Axis")  # Setting label for the x-axis
plt.ylabel("Priyanka's and Arya's Y-Axis")  # Setting label for the y-axis
plt.title("Priyanka's and Arya's")  # Assigning a title to the plot

plt.xlim(0, 6)  # Setting the x-axis limits (lower bound: 0, upper bound: 6)
plt.ylim(0, 12)  # Setting the y-axis limits (lower bound: 0, upper bound: 12)

plt.show()  # Displaying the final plot
```

```python
# Same code, different syntax using Object-Oriented (OO) approach

fig, ax = plt.subplots()  # Create a figure and a set of subplots (ax represents the axes object)

ax.plot(x, y)  # Plot on the axes object instead of calling plt.plot() directly

ax.set_xlabel('x-axis')  # Set label for the x-axis
ax.set_ylabel('y-axis')  # Set label for the y-axis
ax.set_title('OO syntax')  # Assign a title to the plot

plt.show()  # Display the final figure
```

```python
# Alternative OO syntax using plt.figure() and add_subplot()

fig = plt.figure(figsize=(6, 4))  # Creating a figure with a specified size

# Adding subplots manually - grid of 2 rows, 2 columns, selecting each subplot
ax = fig.add_subplot(2, 2, 1)  # First subplot in a 2x2 grid
ax.plot(x, y)  # Plotting the x and y values
ax.set_xlabel('x-axis')  # Setting x-axis label
ax.set_ylabel('y-axis')  # Setting y-axis label
ax.set_title('Using add_subplot()')  # Adding a title to the subplot

ax = fig.add_subplot(2, 2, 2)  # Second subplot in the grid
ax.plot(x, y)  # Plotting again
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Using add_subplot()')

ax = fig.add_subplot(2, 2, 3)  # Third subplot in the grid
ax.plot(x, y)  # Plotting again
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Using add_subplot()')

plt.tight_layout()  # Adjusting layout to prevent overlapping elements
plt.show()  # Displaying the final figure
```

```python
# Simulated data: years and corresponding sales

# Creating an array of years from 2010 to 2020 (inclusive)
a = np.arange(2010, 2021)
print("Years:", a)

# Defining an array of corresponding sales data for each year
b = np.array([100, 120, 130, 160, 180, 200, 240, 260, 280, 300, 320])
print("Sales Data:", b)
```

```python
# Simulated data: age vs. blood pressure

x = np.arange(20, 70, 5)  # Creating age values from 20 to 65 in steps of 5
print("Age:", x)
y = 0.8 * x + 80  # Simulating blood pressure values using a linear equation

# Plot 1: Age vs. Blood Pressure

plt.figure(figsize=(8, 5))  # Setting figure size for better readability
plt.plot(x, y, color='green', marker='o', linestyle='-.', label='BP vs Age')  # Plotting with markers
plt.title('Blood Pressure by Age')  # Adding title to the plot
plt.xlabel("Age (years)")  # Labeling the x-axis
plt.ylabel("Blood Pressure (mm Hg)")  # Labeling the y-axis
plt.grid(True)  # Adding a grid to improve readability
plt.legend()  # Displaying the legend for clarity
plt.tight_layout()  # Adjusting layout to avoid overlap
plt.show()  # Displaying the plot
```

```python
# Plot 2: Year vs. Sales

plt.figure(figsize=(8, 5))  # Creating a figure with size 8x5 for better readability

plt.plot(a, b, color='blue', marker='s', linestyle='--', label='Annual Sales')
# Plotting year (a) vs. sales (b) with blue squares ('s') as markers and dashed ('--') line style

plt.title('Annual Sales Trend')  # Setting the title of the plot
plt.xlabel("Year")  # Labeling the x-axis as "Year"
plt.ylabel("Sales (in $1000s)")  # Labeling the y-axis to indicate sales figures in thousands

plt.grid(True)  # Enabling grid for better readability
plt.legend()  # Displaying the legend to identify the plotted data
plt.tight_layout()  # Adjusting layout to avoid overlapping labels
plt.show()  # Rendering the final plot
```

```python
# Combined Plot

# Create a figure with 1 row and 2 columns to hold both plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Age vs. Blood Pressure
axes[0].plot(x, y, color='green', marker='o', linestyle='-', label='BP vs Age')  # Plotting age vs. BP
axes[0].set_title('Blood Pressure by Age')  # Setting title for the first plot
axes[0].set_xlabel("Age (years)")  # Labeling x-axis
axes[0].set_ylabel("Blood Pressure (mm Hg)")  # Labeling y-axis
axes[0].grid(True)  # Adding grid for better readability
axes[0].legend()  # Displaying the legend for clarity

# Plot 2: Year vs. Sales
axes[1].plot(a, b, color='blue', marker='s', linestyle='--', label='Annual Sales')  # Plotting year vs. sales
axes[1].set_title('Annual Sales Trend')  # Setting title for the second plot
axes[1].set_xlabel("Year")  # Labeling x-axis
axes[1].set_ylabel("Sales (in $1000s)")  # Labeling y-axis
axes[1].grid(True)  # Adding grid for better readability
axes[1].legend()  # Displaying the legend for clarity

plt.tight_layout()  # Adjusting layout to prevent overlap between subplots
plt.show()  # Displaying the final figure
```

```python
# Another version using plt.figure() and add_subplot()

# Create a figure with size 14x5 for better readability
fig = plt.figure(figsize=(14, 5))

# Plot 1: Blood Pressure vs. Age
ax1 = fig.add_subplot(1, 2, 1)  # Adding the first subplot (1 row, 2 columns, first subplot)
ax1.plot(x, y, color='green', marker='o', linestyle='-', label='BP vs Age')  # Plotting age vs. BP
ax1.set_title('Blood Pressure by Age')  # Setting title for the first plot
ax1.set_xlabel("Age (years)")  # Labeling x-axis
ax1.set_ylabel("Blood Pressure (mm Hg)")  # Labeling y-axis
ax1.grid(True)  # Adding grid for better readability
ax1.legend()  # Displaying the legend for clarity

# Plot 2: Year vs. Sales
ax2 = fig.add_subplot(1, 2, 2)  # Adding the second subplot (same row, second column)
ax2.plot(a, b, color='blue', marker='s', linestyle='--', label='Annual Sales')  # Plotting year vs. sales
ax2.set_title('Annual Sales Trend')  # Setting title for the second plot
ax2.set_xlabel("Year")  # Labeling x-axis
ax2.set_ylabel("Sales (in $1000s)")  # Labeling y-axis
ax2.grid(True)  # Adding grid for better readability
ax2.legend()  # Displaying the legend for clarity

# Adjust layout to prevent overlapping labels
plt.tight_layout()
plt.show()
```
