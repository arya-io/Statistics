```python
# Import necessary libraries for plotting and data manipulation
import pandas as pd  # For handling structured data
import matplotlib.pyplot as plt  # For creating visualizations
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/heart.csv")  # Reading the heart dataset into a Pandas DataFrame

# Create a large figure for multiple subplots or detailed visualization
plt.figure(figsize=(18, 16))  # Setting the figure size to 18x16 for clear and spacious plots
```

```python
# 1. Scatter plot - Age vs. Cholesterol

plt.subplot(3, 2, 1)  # First subplot in a 3-row, 2-column grid

plt.scatter(df['age'], df['chol'])
# Creating a scatter plot to visualize cholesterol distribution by age

plt.title("Age vs. Cholesterol")  # Adding a title to the plot
```

```python
# 2. Histogram - Resting Blood Pressure Distribution

plt.subplot(3, 2, 2)  # Creating a subplot (position: 2nd in a 3-row, 2-column grid)

plt.hist(df['trestbps'], bins=20)
# Creating a histogram to show the distribution of resting blood pressure values
# 'bins=20' divides the data into 20 intervals for better granularity

plt.title('Resting BP Distribution')  # Adding a title to the plot
```

```python
# 3. Box Plot - Max Heart Rate

plt.subplot(3, 2, 3)  # Creating a subplot (position: 3rd in a 3-row, 2-column grid)

plt.boxplot(df['thalach'])
# Creating a box plot to display the distribution of maximum heart rate ('thalach')
# Helps identify median, quartiles, and potential outliers

plt.title('Max Heart Rate')  # Adding a title to the plot
```

```python
# 4. Line Plot - Age vs. Cholesterol Trend

plt.subplot(3, 2, 4)  # Creating a subplot (position: 4th in a 3-row, 2-column grid)

df_sorted = df.sort_values(by='age')  # Sorting data by age for a clearer trend line

plt.plot(df_sorted['age'], df_sorted['chol'], linestyle='--', color='b')
# Creating a line plot where age is on the x-axis and cholesterol on the y-axis
# Dashed line style ('--') for trend visibility
# Color set to blue ('b') for differentiation

plt.title('Age vs. Cholesterol Line')  # Adding a title to the plot
```

```python
# 5. Bar Plot - Gender Distribution

plt.subplot(3, 2, 5)  # Creating a subplot (position: 5th in a 3-row, 2-column grid)

df['gender'].value_counts().plot(kind='bar')
# Using Pandas' built-in plotting function to create a bar plot
# 'value_counts()' calculates the frequency of each gender category
# 'kind="bar"' specifies a bar plot format

plt.title('Gender Distribution')  # Adding a title to the plot
```

```python
# 6. Pie Chart - Heart Disease Target Proportion

plt.subplot(3, 2, 6)  # Creating a subplot (position: 6th in a 3-row, 2-column grid)

df['target'].value_counts().plot(kind='pie',
                                 autopct='%1.1f%%',  # Display percentage values with one decimal precision
                                 startangle=90)  # Rotating start angle for better orientation

plt.title('Heart Disease Target Proportion')  # Adding a title to the plot

plt.tight_layout()  # Adjusting layout for optimal spacing
plt.show()  # Displaying the final visualization
```
