```python
# Importing necessary libraries
import pandas as pd  # For handling structured data
import matplotlib.pyplot as plt  # For basic plotting functionalities
import seaborn as sns  # For advanced statistical visualizations
```

```python
# Load the heart dataset into a Pandas DataFrame
df = pd.read_csv('C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/heart.csv')  # Reading the CSV file and storing it in 'df'
```

```python
# Grouping the dataset by 'age' and calculating median values for each group
ages = df.groupby('age').median().reset_index()

# Displaying the first few rows of the resulting DataFrame
print(ages.head())
```

```python
# Line plot using Seaborn to show cholesterol variation by age

sns.lineplot(x='age', y='chol', data=ages, linestyle=':', linewidth=1.7)
# Creating a line plot where 'age' is the x-axis, 'chol' (cholesterol) is the y-axis
# Dashed line style (':') for a distinctive look
# Line width set to 1.7 for better visibility

plt.xlabel("Age")  # Labeling the x-axis
plt.ylabel("Cholesterol")  # Labeling the y-axis
plt.show()  # Displaying the plot
```

```python
# Various plots using Seaborn
# Set Seaborn style for a white grid background

sns.set(style='whitegrid')  # Applying Seaborn's 'whitegrid' style for better readability

plt.figure(figsize=(16, 14))  # Creating a larger figure to accommodate multiple plots
```

```python
# 1. Scatter plot

plt.subplot(3, 3, 1)  # Creating a subplot (position: 1st in a 3x3 grid)

sns.scatterplot(data=df, x='age', y='chol', hue='gender', alpha=0.7)
# Scatter plot showing relationship between age and cholesterol levels
# 'hue' differentiates points based on gender
# 'alpha=0.7' makes points slightly transparent for better overlap visibility

plt.title('Age vs. Cholesterol')  # Adding title to the plot
```

```python
# 2. Histogram with KDE (Kernel Density Estimate)

plt.subplot(3, 3, 2)  # Creating a subplot (position: 2nd in a 3x3 grid)

sns.histplot(df['trestbps'], kde=True, color='blue')
# Creating a histogram to show the distribution of resting blood pressure
# 'kde=True' overlays a density curve to better illustrate distribution shape
# 'color="blue"' sets the bar color

plt.title('Resting BP Distribution')  # Adding a title to the histogram
```

```python
# 3. Box Plot

plt.subplot(3, 2, 3)  # Creating a subplot (position: 3rd in a 3x2 grid)

sns.boxplot(data=df, y='thalach', x='target', palette='Set2')
# Creating a box plot to visualize the distribution of maximum heart rate ('thalach')
# across different target values (typically indicating presence or absence of heart disease)
# 'palette="Set2"' applies a distinct color scheme for better differentiation

plt.title('Max Heart Rate by Target')  # Adding a title to the box plot
```

```python
# 4. Line Plot

plt.subplot(3, 3, 4)  # Creating a subplot (position: 4th in a 3x3 grid)

sns.lineplot(data=df, x='age', y='chol', color='orange')
# Line plot showing cholesterol variation by age
# 'color="orange"' makes the trend easily distinguishable

plt.title('Age vs. Cholesterol Line')  # Adding a title to the plot
```

```python
# 5. Bar Plot (Improved count plot for 'cp' feature)

plt.subplot(3, 3, 5)  # Creating a subplot (position: 5th in a 3x3 grid)

sns.countplot(data=df, x='cp', palette='Set1', hue='target')
# Creating a bar plot to visualize the frequency of different chest pain types ('cp')
# 'palette="Set1"' applies distinct colors to different categories
# 'hue="target"' differentiates bars based on the presence or absence of heart disease

plt.title('Chest Pain Type Distribution By Target')  # Adding a title to the plot
```

```python
# 6. Heatmap (correlation matrix)

plt.subplot(3, 2, 6)  # Creating a subplot (position: 6th in a 3x2 grid)

# Selecting relevant numerical columns for correlation analysis
relevant_columns = ['age', 'chol', 'thalach', 'trestbps', 'oldpeak', 'target']

# Computing the correlation matrix for selected variables
correlation_matrix = df[relevant_columns].corr()

# Creating a heatmap to visualize correlation values
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, linewidth=0.5)
# 'annot=True' displays correlation values within each cell
# 'cmap="coolwarm"' applies a visually appealing color scheme
# 'fmt=".2f"' ensures correlation values are formatted to two decimal places
# 'square=True' keeps cells square-shaped for consistency
# 'linewidth=0.5' adds subtle grid lines for improved readability

plt.title('Heatmap of Selected Variables')  # Adding a title to the heatmap

plt.tight_layout()  # Adjusting layout to prevent overlapping elements
plt.show()  # Displaying the final heatmap
```

```python
# Various plots using Seaborn
# Set seaborn style

sns.set(style='whitegrid')  # Applying Seaborn's white grid background for better readability
plt.figure(figsize=(16, 14))  # Creating a larger figure to accommodate multiple subplots

# 1. Scatter plot - Age vs. Cholesterol
plt.subplot(2, 3, 1)  # First subplot in a 2-row, 3-column grid
sns.scatterplot(data=df, x='age', y='chol', hue='gender', alpha=0.7)
# Scatter plot showing cholesterol levels by age, colored by gender
plt.title('Age vs. Cholesterol')

# 2. Histogram with KDE - Resting BP Distribution
plt.subplot(2, 3, 2)  # Second subplot
sns.histplot(df['trestbps'], kde=True, color='blue')
# Histogram of resting blood pressure values with a KDE overlay
plt.title('Resting BP Distribution')

# 3. Box Plot - Maximum Heart Rate by Target
plt.subplot(2, 3, 3)  # Third subplot
sns.boxplot(data=df, y='thalach', x='target', palette='Set2')
# Box plot comparing heart rate distribution across target values
plt.title('Max Heart Rate by Target')

# 4. Line Plot - Age vs. Cholesterol Trend
plt.subplot(2, 3, 4)  # Fourth subplot
sns.lineplot(data=df, x='age', y='chol', color='orange')
# Line plot showing cholesterol trend by age
plt.title('Age vs. Cholesterol Line')

# 5. Bar Plot - Chest Pain Type Distribution by Target
plt.subplot(2, 3, 5)  # Fifth subplot
sns.countplot(data=df, x='cp', palette='Set1', hue='target')
# Count plot showing distribution of chest pain types, categorized by target
plt.title('Chest Pain Type Distribution By Target')

# 6. Heatmap - Correlation Matrix of Selected Features
plt.subplot(2, 3, 6)  # Sixth subplot
relevant_columns = ['age', 'chol', 'thalach', 'trestbps', 'oldpeak', 'target']
correlation_matrix = df[relevant_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, linewidth=0.5)
# Heatmap visualizing correlations among selected numerical variables
plt.title('Heatmap of Selected Variables')

plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()  # Display the final figure
```
