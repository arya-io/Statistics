***Covariance & correlation***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/weight-height.csv")
```

```python
print(df.head())
```

```python
# Separating male and female data (assuming you have more data)
df_male = df[df['Gender'] == 'Male']
df_female = df[df['Gender'] == 'Female']  # Adjust this when female data is available

# Display previews of each DataFrame
print("Preview of Male Data:")
print(df_male.head())

print("Preview of Female Data:")
print(df_female.head())  # Currently empty, needs female data
```

```python
# Step 1: Calculate correlation for the entire dataset
# The correlation matrix shows the relationship between 'Height' and 'Weight' across all records
corr_overall = df[['Height', 'Weight']].corr()

# Step 2: Calculate correlation specifically for male entries
# This helps analyze how height and weight are related for males only
corr_male = df_male[['Height', 'Weight']].corr()

# Step 3: Calculate correlation specifically for female entries
# This helps analyze how height and weight are related for females only
corr_female = df_female[['Height', 'Weight']].corr()

# Step 4: Display results for easier analysis
print("Overall Correlation Matrix:")
print(corr_overall)

print("\nMale Correlation Matrix:")
print(corr_male)

print("\nFemale Correlation Matrix:")
print(corr_female)
```

```python
# Step 1: Calculate covariance for the entire dataset
# Covariance helps determine how two variables (Height & Weight) change together
cov_overall = df[['Height', 'Weight']].cov()

# Step 2: Calculate covariance specifically for male entries
# Helps analyze the relationship between height and weight among males
cov_male = df_male[['Height', 'Weight']].cov()

# Step 3: Calculate covariance specifically for female entries
# Helps analyze the relationship between height and weight among females
cov_female = df_female[['Height', 'Weight']].cov()

# Step 4: Display results for easier analysis
print("Overall Covariance Matrix:")
print(cov_overall)

print("\nMale Covariance Matrix:")
print(cov_male)

print("\nFemale Covariance Matrix:")
print(cov_female)
```

```python
# Step 1: Display overall correlation between Height and Weight
print("\nOverall correlation between Height and Weight:")
print(corr_overall)  # Shows how height and weight are related across all records

# Step 2: Display correlation for males
print("\nMale correlation between Height and Weight:")
print(corr_male)  # Shows how height and weight are related among males

# Step 3: Display correlation for females
print("\nFemale correlation between Height and Weight:")
print(corr_female)  # Shows how height and weight are related among females

# Step 4: Display overall covariance between Height and Weight
print("\nOverall Covariance between Height and Weight:")
print(cov_overall)  # Measures how height and weight vary together across all records

# Step 5: Display covariance for males
print("\nMale Covariance between Height and Weight:")
print(cov_male)  # Measures how height and weight vary together among males

# Step 6: Display covariance for females
print("\nFemale Covariance between Height and Weight:")
print(cov_female)  # Measures how height and weight vary together among females
```

```python
import matplotlib.pyplot as plt  # Import Matplotlib for data visualization

# Step 1: Create a figure with three subplots in a row
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # Creates three scatter plots side by side

# Step 2: Scatter plot for overall data
# Visualizes the relationship between height and weight across all individuals
axes[0].scatter(df['Height'], df['Weight'], color='blue', alpha=0.5)
axes[0].set_title(f'Overall correlation: {corr_overall.loc["Height", "Weight"]:.2f}')  # Displays correlation value
axes[0].set_xlabel('Height')
axes[0].set_ylabel('Weight')
axes[0].grid(True)  # Adds grid for better readability

# Step 3: Scatter plot for male data
# Shows correlation specifically among males
axes[1].scatter(df_male['Height'], df_male['Weight'], color='green', alpha=0.5)
axes[1].set_title(f'Male correlation: {corr_male.loc["Height", "Weight"]:.2f}')  # Displays correlation value
axes[1].set_xlabel('Height')
axes[1].set_ylabel('Weight')
axes[1].grid(True)

# Step 4: Scatter plot for female data
# Displays correlation for female height vs weight
axes[2].scatter(df_female['Height'], df_female['Weight'], color='pink', alpha=0.5)
axes[2].set_title(f'Female correlation: {corr_female.loc["Height", "Weight"]:.2f}')  # Displays correlation value
axes[2].set_xlabel('Height')
axes[2].set_ylabel('Weight')
axes[2].grid(True)

# Step 5: Adjust layout to prevent overlapping elements
plt.tight_layout()

# Step 6: Display the scatter plots
plt.show()
```

```python
# Step 1: Set up the figure size for better readability
plt.figure(figsize=(6,4))  # Defines the plot size to ensure a clear and structured visualization

# Step 2: Create a heatmap to visualize the correlation matrix
# A heatmap helps show relationships between different variables using color intensity
sns.heatmap(corr_overall, annot=True, fmt=".2f", cmap="coolwarm")  # 'annot=True' adds values in the heatmap

# Step 3: Add a title for better context
plt.title('Overall Correlation Heatmap')

# Step 4: Adjust labels for better readability
plt.xticks(rotation=45)  # Rotates x-axis labels to prevent overlap
plt.yticks(rotation=0)  # Keeps y-axis labels horizontal for easy viewing

# Step 5: Apply tight layout to optimize spacing
plt.tight_layout()  # Ensures proper spacing of elements to avoid overlapping

# Step 6: Display the heatmap
plt.show()
```
