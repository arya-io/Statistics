```python
import pandas as pd  # Importing Pandas for data handling

# Step 1: Load the dataset into a Pandas DataFrame
# 'pd.read_csv()' reads the CSV file and stores it in a structured format
file_path = 'C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/euro.csv'
df = pd.read_csv(file_path)

# Step 2: Remove any rows with missing values (NaN)
# The 'dropna()' function ensures clean data by removing incomplete records
df = df.dropna()

# Step 3: Convert INR and USD columns to float data type
# This ensures numerical consistency for calculations and analysis
df['INR'] = df['INR'].astype(float)
df['USD'] = df['USD'].astype(float)

# Step 4: Display a preview of the cleaned dataset
print("Preview of the cleaned dataset:")
print(df.head())  # Prints the first five rows for quick inspection
```

```python
df
```

```python
# Step 1: Calculate the correlation between INR and USD exchange rates
# Pearson correlation helps understand how strongly INR and USD are related
correlation_coefficient = df['INR'].corr(df['USD'])

# Step 2: Display the result with context
print("Correlation Coefficient between INR and USD exchange rates:", correlation_coefficient)
```

```python
import matplotlib.pyplot as plt  # Import Matplotlib for visualization

# Step 1: Create a scatter plot to show the relationship between INR and USD exchange rates
plt.scatter(df['INR'], df['USD'], alpha=0.6, color='blue')  # Added transparency (alpha) for better clarity

# Step 2: Label the axes for better readability
plt.xlabel('Indian Rupee (INR)')  # X-axis represents INR values
plt.ylabel('US Dollar (USD)')  # Y-axis represents USD values

# Step 3: Add a title to provide context and display the correlation value
plt.title(f'Scatter Plot (Correlation: {correlation_coefficient:.2f})')

# Step 4: Apply grid lines to enhance readability
plt.grid(True)  # Helps in aligning points visually

# Step 5: Display the plot
plt.show()
```

```python
import seaborn as sns  # Importing Seaborn for statistical visualization
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Step 1: Create a joint plot to visualize the relationship between INR and USD
# A jointplot provides both a scatter plot and marginal histograms for a comprehensive view
sns.jointplot(data=df, x='INR', y='USD', kind='scatter', color='blue')

# Step 2: Show the plot
plt.show()
```
