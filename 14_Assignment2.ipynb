1) For diabetes.csv dataset, find the conditional probability that a person's Glucose levels are > 110 given that the person's BMI >= 25

```python
# Marginal and Joint(Unconditional) probability
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/diabetes.csv")
```

```python
df
```

```python
# Step 1: Select only relevant columns (Glucose & BMI)
df = df[['Glucose', 'BMI']]

# Step 2: Remove rows with missing values to ensure clean data
df.dropna(inplace=True)

# Step 3: Print dataset preview after cleaning
print("Preview of the cleaned dataset:")
print(df.head())
```

```python
# Step 1: Count the total number of rows in the cleaned DataFrame
total = len(df)

# Step 2: Print the result for quick verification
print(f"Total rows after filtering and cleaning: {total}")
```

```python
# Step 1: Calculate the marginal probability of Glucose > 110
p_glucose = len(df[df['Glucose'] > 110]) / total  # Proportion of high Glucose levels

# Step 2: Calculate the marginal probability of BMI â‰¥ 25
p_bmi = len(df[df['BMI'] >= 25]) / total  # Proportion of BMI at or above 25

# Step 3: Print probabilities with proper formatting
print(f"P(Glucose > 110): {p_glucose:.3f}")  # Rounded to 3 decimal places
print(f"P(BMI >= 25): {p_bmi:.3f}")  # Rounded to 3 decimal places
```

```python
# Step 1: Calculate the joint probability of Glucose > 110 AND BMI â‰¥ 25
p_glucose_bmi = len(df[(df['Glucose'] > 110) & (df['BMI'] >= 25)]) / total

# Step 2: Print the result with proper formatting
print(f"P(Glucose > 110 and BMI â‰¥ 25): {p_glucose_bmi:.3f}")  # Rounded to 3 decimal places
```

```python
# Step 1: Compute the conditional probability P(Glucose > 110 | BMI â‰¥ 25)
p_glucose_given_bmi = p_glucose_bmi / p_bmi

# Step 2: Print the result with proper formatting
print("\nConditional Probabilities:")
print(f"P(Glucose > 110 | BMI >= 25): {p_glucose_given_bmi:.3f}")  # Rounded to 3 decimal places
```

Q2. For the combined Marathon datasets ... Find the conditional probability of Finish within 3 hours given that pace < 8 minutes per mile

```python
df_2015 = pd.read_csv("C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/marathon_results_2015.csv")
df_2016 = pd.read_csv("C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/marathon_results_2016.csv")
df_2017 = pd.read_csv("C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/marathon_results_2017.csv")
```

```python
# Concatenate the datasets into a single DataFrame

df = pd.concat([df_2015, df_2016, df_2017])
df
```

```python
# Step 1: Split 'Official Time' into hours, minutes, and seconds
df[['of_Hours', 'of_Minutes', 'of_Seconds']] = df['Official Time'].str.split(':', expand=True)

# Step 2: Split 'Pace' into hours, minutes, and seconds
df[['pace_Hours', 'pace_Minutes', 'pace_Seconds']] = df['Pace'].str.split(':', expand=True)

# Step 3: Convert extracted values to numeric type for calculations
df[['of_Hours', 'of_Minutes', 'of_Seconds', 'pace_Hours', 'pace_Minutes', 'pace_Seconds']] = df[
    ['of_Hours', 'of_Minutes', 'of_Seconds', 'pace_Hours', 'pace_Minutes', 'pace_Seconds']
].astype(int)

# Step 4: Display the modified DataFrame
df.head()  # Shows a preview of the updated dataset
```

```python
df['of_Hours'] = pd.to_numeric(df['of_Hours'])
df['of_Minutes'] = pd.to_numeric(df['of_Minutes'])
df['of_Seconds'] = pd.to_numeric(df['of_Seconds'])
df['pace_Hours'] = pd.to_numeric(df['pace_Hours'])
df['pace_Minutes'] = pd.to_numeric(df['pace_Minutes'])
df['pace_Seconds'] = pd.to_numeric(df['pace_Seconds'])

# # Convert time-related columns to numeric for calculations
# time_cols = ['of_Hours', 'of_Minutes', 'of_Seconds', 'pace_Hours', 'pace_Minutes', 'pace_Seconds']
# df[time_cols] = df[time_cols].apply(pd.to_numeric)

# # Print a preview to verify successful conversion
# print(df.dtypes)  # Displays column data types
```

```python
# Step 1: Convert 'Official Time' to total seconds
df["of_Total"] = df["of_Hours"] * 3600 + df["of_Minutes"] * 60 + df["of_Seconds"]

# Step 2: Convert 'Pace' to total seconds
df["pace_Total"] = df["pace_Hours"] * 3600 + df["pace_Minutes"] * 60 + df["pace_Seconds"]

# Step 3: Display a preview for verification
print(df[['of_Total', 'pace_Total']].head())  # Shows the first few rows
```

```python
df.head()
```

```python
# Step 1: Select only the relevant columns for analysis
df = df[['of_Total', 'pace_Total']]

# Step 2: Remove rows with missing values to ensure clean data
df.dropna(inplace=True)

# Step 3: Display the cleaned dataset preview
print("Preview of the refined dataset:")
print(df.head())  # Shows the first few rows for quick inspection
```

```python
# Step 1: Count the total number of rows in the DataFrame
total = len(df)

# Step 2: Print the result for verification
print(f"Total rows in the DataFrame: {total}")
```

```python
# Step 1: Compute the marginal probability of finishing within 3 hours
p_of_total = len(df[df['of_Total'] <= (3*60*60)]) / total  # Total seconds in 3 hours

# Step 2: Compute the marginal probability of pace being under 8 minutes/mile
p_pace_total = len(df[df['pace_Total'] < (8*60)]) / total  # Total seconds in 8 minutes

# Step 3: Print the results with proper formatting
print(f"P(of_Total â‰¤ 3 Hours): {p_of_total:.3f}")  # Rounded to 3 decimal places
print(f"P(pace_Total < 8 Minutes): {p_pace_total:.3f}")  # Rounded to 3 decimal places
```

```python
# Step 1: Compute the joint probability of finishing within 3 hours AND pace under 8 minutes per mile
p_official_pace = len(df[(df['of_Total'] <= (3 * 60 * 60)) & (df['pace_Total'] < (8 * 60))]) / total

# Step 2: Print the result with proper formatting
print(f"P(of_Total â‰¤ 3 Hours and pace_Total < 8 Minutes): {p_official_pace:.3f}")
```

```python
# Step 1: Compute the conditional probability P(of_Total â‰¤ 3 Hours | pace_Total < 8 Minutes)
p_total_given_pace = p_official_pace / p_pace_total  # Uses joint probability & marginal probability

# Step 2: Print the result with proper formatting
print("\nConditional Probabilities:")
print(f"P(of_Total â‰¤ 3 Hours | pace_Total < 8 Minutes): {p_total_given_pace:.3f}")  # Rounded to 3 decimal places
```

Q3). Find the conditional probability of winning a nobel prize in physics given the laureate is female.

```python
df = pd.read_csv("C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/nobel.csv")
```

```python
N = len(df)
print(f"Total rows in the DataFrame: {N}")
```

```python
df
```

```python
# Step 1: Remove rows where 'gender' column has missing values
df.dropna(subset=['gender'], inplace=True)

# Step 2: Display dataset preview to verify changes
print("Preview after dropping missing gender values:")
print(df.head())
```

```python
# Step 1: Count the total number of rows in the DataFrame
N = len(df)

# Step 2: Print the result for verification
print(f"Total rows in the DataFrame: {N}")
```

```python
df.head()
```

```python
# Step 1: Count the number of entries where 'category' is 'Physics'
A = len(df[df['category'] == 'Physics'])

# Step 2: Print the result for verification
print(f"Total entries in the 'Physics' category: {A}")
```

```python
print(df['gender'].unique())  # Shows distinct gender categories
```

```python
# Step 1: Count the number of entries where 'gender' is 'female'
B = len(df[df['gender'] == 'female'])

# Step 2: Print the result for verification
print(f"Total number of female entries: {B}")
```

```python
# Step 1: Count the number of entries where 'category' is 'Physics' AND 'gender' is 'female'
A_and_B = len(df[(df['category'] == 'Physics') & (df['gender'] == 'female')])

# Step 2: Print the result for verification
print(f"Total number of female entries in the 'Physics' category: {A_and_B}")
```

```python
# Step 1: Calculate marginal probabilities
P_A = A / N  # Probability of category being 'Physics'
P_B = B / N  # Probability of gender being 'female'

# Step 2: Calculate conditional probability P(female | Physics)
P_B_given_A = A_and_B / A  # Probability of being female, given the category is 'Physics'

# Step 3: Print rounded values for readability
print(round(P_A, 2))  # Marginal probability of 'Physics'
print(round(P_B, 2))  # Marginal probability of 'female'
print(round(P_B_given_A, 2))  # Conditional probability P(female | Physics)
```

```python
# Step 1: Compute conditional probability P(Physics | Female) using Bayes' Theorem
P_A_given_B = (P_B_given_A * P_A) / P_B

# Step 2: Print the result for verification
print(f"P(Physics | Female): {round(P_A_given_B, 2)}")
```

```python
# Step 1: Calculate P(A | B) = P(Physics | Female)
P_A_given_B = A_and_B / B  # Probability of 'Physics' given that gender is 'female'

# Step 2: Print rounded result
print(round(P_A_given_B, 2))
```
