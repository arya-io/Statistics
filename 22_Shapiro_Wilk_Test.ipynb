```python
# Import necessary libraries
import pandas as pd  # For data manipulation
import matplotlib.pyplot as plt  # For plotting QQ plot
import scipy.stats as stats  # For statistical tests
import numpy as np  # For numerical operations

# The Shapiro-Wilk test and QQ plot are used to check whether data follows a normal distribution.
# - Shapiro-Wilk Test: Provides a statistical test for normality.
# - QQ Plot: Helps visualize if data fits a normal distribution.
```

```python
# Load the dataset from the specified path
# Ensure the file path is correct; update it if needed
data = pd.read_csv('C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/tips.csv')

# Display the summary information of the dataset
# This provides details on column names, data types, and non-null values
data.info()
```

```python
# Extract the 'total_bill' column from the dataset
# This creates a Pandas Series containing only the total bill amounts
total_bill_series = data['total_bill']

# Now, 'total_bill_series' holds the bill values, which can be analyzed further.
# You can check its basic statistics using:
print(total_bill_series.describe())  # Provides summary statistics (mean, min, max, etc.)
```

```python
# Normality test with Shapiro-Wilk test

shapiro_test = stats.shapiro(total_bill_series)
print("Test statistic (W):", shapiro_test.pvalue)

if shapiro_test.pvalue > 0.05:
    print("Data is normally distributed")
else:
    print("Data is not normally distributed")
```

```python
# Generate a QQ Plot to visually assess whether 'total_bill' follows a normal distribution
# QQ Plot compares the quantiles of our dataset with a theoretical normal distribution

stats.probplot(total_bill_series, dist="norm", plot=plt)  # Creates the QQ plot

# Add a title to the plot
plt.title('QQ Plot of Total Bill')

# Label the x-axis to indicate theoretical quantiles from a normal distribution
plt.xlabel('Theoretical Quantiles (Normal Distribution)')

# Label the y-axis to show the actual ordered values from the dataset
plt.ylabel('Ordered values (Total Bill)')

# Enable grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Load the rainfall dataset from the specified path
# Ensure the file path is correct before running the code
data = pd.read_csv('C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/jersey-rainfall-1984-to-2021.csv')

# Display summary information about the dataset
# This provides details on column names, data types, and missing values
data.info()
```

```python
# Extract the 'YearTotalmm' column from the dataset
# This creates a Pandas Series containing the total annual rainfall measurements
total_bill_series = data['YearTotalmm']

# Now, 'total_bill_series' holds the yearly rainfall data, which can be analyzed further.
# You can check basic statistics using:
print(total_bill_series.describe())  # Provides summary statistics (mean, min, max, etc.)
```

```python
# Perform the Shapiro-Wilk normality test
# This test checks if the 'total_bill_series' data follows a normal distribution
shapiro_test = stats.shapiro(total_bill_series)

# Print the test statistic's p-value
# The p-value helps determine whether the data follows a normal distribution
print("Test statistic (W):", shapiro_test.pvalue)

# Interpretation:
# - If p-value > 0.05: Fail to reject the null hypothesis, data is likely normally distributed
# - If p-value â‰¤ 0.05: Reject the null hypothesis, data is likely not normally distributed
if shapiro_test.pvalue > 0.05:
    print("Data is normally distributed")
else:
    print("Data is not normally distributed")
```

```python
# Generate a QQ Plot to visually assess whether 'total_bill_series' follows a normal distribution
# QQ Plot compares the quantiles of our dataset with a theoretical normal distribution

stats.probplot(total_bill_series, dist="norm", plot=plt)  # Creates the QQ plot

# Add a title to the plot
plt.title('QQ Plot of Total Bill')

# Label the x-axis to indicate theoretical quantiles from a normal distribution
plt.xlabel('Theoretical Quantiles (Normal Distribution)')

# Label the y-axis to show the actual ordered values from the dataset
plt.ylabel('Ordered values (Total Bill)')

# Enable grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```
