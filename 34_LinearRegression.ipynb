```python
# Import necessary libraries
import pandas as pd  # For handling datasets

# Placeholder: Load your dataset (Update the file path when needed)
# df = pd.read_csv("path_to_your_dataset.csv")

# Display the first few rows of the dataset (Uncomment after loading the dataset)
# print(df.head())
```

```python
# Step 1: Load the dataset and identify feature (X) and label (y)

# Load the salary dataset from the specified path
df = pd.read_csv('C:/Users/dbda.STUDENTSDC/Music/LabPractice/Notebooks/Datasets/salary.csv')

# Define the feature (independent variable) and label (dependent variable)
X = df[['YearsExperience']]  # Predictor: Years of Experience
y = df['Salary']  # Target variable: Salary

# Display the first few rows to verify the structure
print(df.head())
```

```python
# Step 2: Split data into training and testing sets

from sklearn.model_selection import train_test_split

# Splitting the dataset: 80% training data, 20% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

# Print dataset sizes for verification
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
# Print the test set feature and label values for verification
print("--- Test Set Data ---")
print("X_test (Years of Experience):\n", X_test)  # Displays independent variable values
print("\ny_test (Salary):\n", y_test)  # Displays dependent variable values
print("----------------\n")
```

```python
# Step 3: Fit Simple Linear Regression to Training Data

from sklearn.linear_model import LinearRegression

# Initialize the linear regression model
model = LinearRegression()

# Fit the model using training data (X_train, y_train)
model.fit(X_train, y_train)

# Print model coefficients for interpretation
print("Intercept:", model.intercept_)  # The constant term
print("Slope:", model.coef_[0])  # The weight assigned to 'YearsExperience'
```

```python
# Step 4: Make Predictions on the Test Set

# Predict the salary values based on the test set features (Years of Experience)
y_pred = model.predict(X_test)

# Print the predicted values for the test set
print("Predicted values on test set:\n", y_pred)
```

```python
# Step 5: Visualize training set results

import matplotlib.pyplot as plt  # Importing Matplotlib for visualization

# Scatter plot of actual training data points
plt.scatter(X_train, y_train, color='red', label="Actual Data")

# Regression line based on model predictions
plt.plot(X_train, model.predict(X_train), color='blue', label="Regression Line")

# Set title and axis labels for clarity
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

# Add legend for better readability
plt.legend()

# Show the plot
plt.show()
```

```python
# Step 6: Visualize Test Set Results

import matplotlib.pyplot as plt  # Import visualization library

# Scatter plot for actual test set results
plt.scatter(X_test, y_test, color='red', label="Actual")  

# Line plot for predicted values based on the trained model
plt.plot(X_train, model.predict(X_train), color='blue', label="Model Prediction")  

# Add title and labels
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

# Display legend for clarity
plt.legend()

# Show the plot
plt.show()
```

```python
# Step 7: Make new predictions for unseen experience values

# Create a DataFrame with new experience values for prediction
new_X_values_df = pd.DataFrame({'YearsExperience': [1,3,5,8,12,15,20,25]})

# Predict salary using the trained model
new_y_pred = model.predict(new_X_values_df)

# Print predictions in a structured format
print('\nPredictions for new Experience values (1, 3, 5, 8, 12, 15, 20, 25):\n', new_y_pred)
```

```python
# Step 8: Model Evaluation

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Compute evaluation metrics
mae = mean_absolute_error(y_test, y_pred)  # Measures average absolute error
mse = mean_squared_error(y_test, y_pred)  # Measures squared error magnitude
r2 = r2_score(y_test, y_pred)  # Determines how well the model explains variance in the data

# Print evaluation results
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2 Score):", r2)
```
